```python
#Write a Python program to generate a random variable and display its value

import random
random_variable = random.randint(1, 100)
print("Random Variable:", random_variable)
``"
# import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import discrete_uniform

# Define the parameters of the discrete uniform distribution
a = 1  # Lower bound
b = 10  # Upper bound

# Generate the discrete uniform distribution
x = np.arange(a, b + 1)  # Possible values
pmf = discrete_uniform.pmf(x, a, b)  # Probability mass function

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.plot(x, pmf, 'o-')
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(True)
plt.show()


# > Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

import numpy as np

def bernoulli_pdf(x, p):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1)
    p (float): The probability of success (0 ≤ p ≤ 1)

    Returns:
    float: The probability of the given value
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1")

    if x == 1:
        return p
    else:
        return 1 - p

# Example usage:
x = 1  # Value of the random variable
p = 0.7  # Probability of success

probability = bernoulli_pdf(x, p)
print(f"P(X = {x}) = {probability}")


# Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Simulate the binomial distribution
num_simulations = 10000
simulated_data = np.random.binomial(n, p, num_simulations)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(simulated_data, bins=11, edgecolor='black', align='left', rwidth=0.8)
plt.title('Binomial Distribution Histogram')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(np.arange(n + 1))
plt.grid(True)
plt.show()

# Optional: Plot the theoretical binomial distribution
x = np.arange(n + 1)
theoretical_probabilities = binom.pmf(x, n, p)
plt.figure(figsize=(8, 6))
plt.plot(x, theoretical_probabilities, 'o-', label='Theoretical Distribution')
plt.title('Theoretical Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(n + 1))
plt.grid(True)
plt.legend()
plt.show()

# Create a Poisson distribution and visualize it using Python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the parameter of the Poisson distribution (lambda)
lambda_param = 5

# Create a range of x values from 0 to 15
x = np.arange(0, 16)

# Calculate the probability mass function (PMF) of the Poisson distribution
pmf = poisson.pmf(x, lambda_param)

# Create a bar plot of the PMF
plt.figure(figsize=(8, 6))
plt.bar(x, pmf)
plt.title('Poisson Distribution (λ = {})'.format(lambda_param))
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Optional: Plot the cumulative distribution function (CDF)
cdf = poisson.cdf(x, lambda_param)
plt.figure(figsize=(8, 6))
plt.plot(x, cdf, 'o-')
plt.title('Poisson Distribution CDF (λ = {})'.format(lambda_param))
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()


# Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the parameters of the discrete uniform distribution
a = 1  # Lower bound
b = 10  # Upper bound

# Create a range of x values from a to b
x = np.arange(a, b + 1)

# Calculate the cumulative distribution function (CDF)
cdf = np.cumsum(np.ones_like(x)) / len(x)

# Create a plot of the CDF
plt.figure(figsize=(8, 6))
plt.plot(x, cdf, 'o-')
plt.title('Discrete Uniform Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()


# Generate a continuous uniform distribution using NumPy and visualize it
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the continuous uniform distribution
a = 0  # Lower bound
b = 10  # Upper bound

# Generate the continuous uniform distribution
x = np.linspace(a, b, 1000)  # Generate 1000 evenly spaced values over the range [a, b]
y = uniform.pdf(x, loc=a, scale=b - a)  # Calculate the probability density function (PDF)

# Create a plot of the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Continuous Uniform Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Optional: Generate random samples from the distribution
samples = np.random.uniform(a, b, 1000)  # Generate 1000 random samples from the distribution
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')
plt.title('Continuous Uniform Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# Simulate data from a normal distribution and plot its histogram

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the normal distribution
mean = 5  # Mean
std_dev = 2  # Standard deviation

# Simulate data from the normal distribution
data = np.random.normal(mean, std_dev, 1000)  # Generate 1000 random samples

# Create a histogram of the data
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the probability density function (PDF) of the normal distribution
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, 'r-', lw=2, label='Normal Distribution PDF')

plt.title('Simulated Normal Distribution Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()


#Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (numpy array): Input dataset
    """
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev

    # Create a histogram of the Z-scores
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='g')

    # Plot the standard normal distribution curve
    x = np.linspace(-3, 3, 100)
    y = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)
    plt.plot(x, y, 'r-', lw=2, label='Standard Normal Distribution')

    plt.title('Z-Scores Histogram')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()


# Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the exponential distribution
lambda_param = 1  # Rate parameter

# Define the number of samples and sample size
num_samples = 1000
sample_size = 30

# Generate samples from the exponential distribution
samples = np.random.exponential(scale=1/lambda_param, size=(num_samples, sample_size))

# Calculate the sample means
sample_means = np.mean(samples, axis=1)

# Plot the histogram of the sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
x = np.linspace(np.mean(sample_means) - 3 * np.std(sample_means), np.mean(sample_means) + 3 * np.std(sample_means), 100)
y = np.exp(-((x - np.mean(sample_means)) / np.std(sample_means))**2 / 2) / (np.std(sample_means) * np.sqrt(2 * np.pi))
plt.plot(x, y, 'r-', lw=2, label='Normal Distribution')

plt.title('Central Limit Theorem (CLT) Demonstration')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()


# Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the normal distribution
mean = 5  # Mean
std_dev = 2  # Standard deviation

# Define the number of samples and sample size
num_samples = 1000
sample_sizes = [5, 10, 20, 50]  # Different sample sizes to demonstrate the CLT

# Create a figure with multiple subplots
fig, axs = plt.subplots(len(sample_sizes), figsize=(8, 6 * len(sample_sizes)))

# Simulate samples and plot histograms for each sample size
for i, sample_size in enumerate(sample_sizes):
    samples = np.random.normal(mean, std_dev, (num_samples, sample_size))
    sample_means = np.mean(samples, axis=1)
    
    axs[i].hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
    
    # Plot the normal distribution curve
    x = np.linspace(np.mean(sample_means) - 3 * np.std(sample_means), np.mean(sample_means) + 3 * np.std(sample_means), 100)
    y = norm.pdf(x, np.mean(sample_means), np.std(sample_means))
    axs[i].plot(x, y, 'r-', lw=2, label='Normal Distribution')
    
    axs[i].set_title(f'Sample Size: {sample_size}')
    axs[i].set_xlabel('Sample Mean')
    axs[i].set_ylabel('Frequency')
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()


# Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)5

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Calculate and plot the standard normal distribution (mean = 0, std = 1).
    """
    # Define the range of x values
    x = np.linspace(-4, 4, 100)

    # Calculate the probability density function (PDF) of the standard normal distribution
    y = norm.pdf(x, loc=0, scale=1)

    # Create a plot of the standard normal distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x, y)
    plt.title('Standard Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()


# Generate random variables and calculate their corresponding probabilities using the binomial distribution

import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate random variables (number of successes)
x = np.arange(0, n + 1)

# Calculate the probabilities using the binomial distribution
probabilities = binom.pmf(x, n, p)

# Print the probabilities
print("Probabilities:")
for i in range(len(x)):
    print(f"P(X = {x[i]}) = {probabilities[i]:.4f}")

# Plot the probabilities
plt.figure(figsize=(8, 6))
plt.bar(x, probabilities)
plt.title('Binomial Distribution Probabilities')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Generate random samples from the binomial distribution
samples = np.random.binomial(n, p, 1000)

# Plot a histogram of the samples
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=n + 1, edgecolor='black', align='left', rwidth=0.8, density=True)
plt.title('Binomial Distribution Samples')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation of the dataset
mean = 5
std_dev = 2

# Define the data point for which to calculate the Z-score
data_point = 8

# Calculate the Z-score
z_score = (data_point - mean) / std_dev
print(f"Z-score: {z_score:.2f}")

# Create a range of x values for the standard normal distribution
x = np.linspace(-3, 3, 100)

# Calculate the probability density function (PDF) of the standard normal distribution
y = norm.pdf(x)

# Create a plot of the standard normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Standard Normal Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.grid(True)

# Plot the Z-score as a vertical line
plt.axvline(z_score, color='r', linestyle='--', label=f'Z-Score: {z_score:.2f}')
plt.legend()
plt.show()

# Calculate the probability of observing a value greater than the data point
probability = 1 - norm.cdf(z_score)
print(f"Probability of observing a value greater than {data_point}: {probability:.4f}")


# Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample dataset
data = np.array([23, 11, 19, 24, 17, 22, 20, 25, 21, 18])

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

# Define the null and alternative hypotheses
null_hypothesis = "The population mean is equal to 20."
alternative_hypothesis = "The population mean is not equal to 20."

# Define the significance level (alpha)
alpha = 0.05

# Calculate the Z-statistic
z_statistic = (sample_mean - 20) / (sample_std_dev / np.sqrt(len(data)))

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

# Print the results
print("Null Hypothesis:", null_hypothesis)
print("Alternative Hypothesis:", alternative_hypothesis)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
print("Z-Statistic:", z_statistic)
print("p-Value:", p_value)

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Plot the distribution of the sample means
sample_means = np.array([np.mean(np.random.choice(data, size=len(data), replace=True)) for _ in range(1000)])
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.legend()
plt.show()


# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Create a sample dataset
np.random.seed(0)
data = np.random.normal(loc=5, scale=2, size=100)

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)

# Define the confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate the standard error
standard_error = sample_std_dev / np.sqrt(len(data))

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=len(data) - 1) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
print("Confidence Interval (95%):", confidence_interval)

# Plot the dataset and the confidence interval
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.axvline(confidence_interval[0], color='b', linestyle='--', label='95% Confidence Interval')
plt.axvline(confidence_interval[1], color='b', linestyle='--')
plt.legend()
plt.show()


# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the normal distribution
mean = 5
std_dev = 2

# Generate data from the normal distribution
data = np.random.normal(mean, std_dev, 100)

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)

# Define the confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate the standard error
standard_error = sample_std_dev / np.sqrt(len(data))

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=len(data) - 1) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
print("Confidence Interval (95%):", confidence_interval)

# Interpret the confidence interval
print("Interpretation: We can be 95% confident that the true population mean lies between {:.2f} and {:.2f}.".format(confidence_interval[0], confidence_interval[1]))

# Plot the data and the confidence interval
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.axvline(confidence_interval[0], color='b', linestyle='--', label='95% Confidence Interval')
plt.axvline(confidence_interval[1], color='b', linestyle='--')
plt.legend()
plt.show()


# Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the normal distribution
mean = 5
std_dev = 2

# Create a range of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the probability density function (PDF) of the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Create a plot of the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf)
plt.title('Normal Distribution Probability Density Function (PDF)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Calculate the cumulative distribution function (CDF) of the normal distribution
cdf = norm.cdf(x, loc=mean, scale=std_dev)

# Create a plot of the CDF
plt.figure(figsize=(8, 6))
plt.plot(x, cdf)
plt.title('Normal Distribution Cumulative Distribution Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# Generate random samples from the normal distribution
samples = np.random.normal(mean, std_dev, 1000)

# Create a histogram of the samples
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution Random Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the parameter of the Poisson distribution (lambda)
lambda_param = 5

# Create a range of x values (number of occurrences)
x = np.arange(0, 20)

# Calculate the probability mass function (PMF) of the Poisson distribution
pmf = poisson.pmf(x, lambda_param)

# Calculate the cumulative distribution function (CDF) of the Poisson distribution
cdf = poisson.cdf(x, lambda_param)

# Print the CDF values
print("Cumulative Distribution Function (CDF) values:")
for i in range(len(x)):
    print(f"P(X ≤ {x[i]}) = {cdf[i]:.4f}")

# Create a plot of the PMF and CDF
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(x, pmf)
plt.title('Poisson Distribution Probability Mass Function (PMF)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, marker='o')
plt.title('Poisson Distribution Cumulative Distribution Function (CDF)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()


# Simulate a random variable using a continuous uniform distribution and calculate its expected value

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the continuous uniform distribution
lower_bound = 0
upper_bound = 10

# Simulate a random variable using the continuous uniform distribution
random_variable = np.random.uniform(lower_bound, upper_bound, 1000)

# Calculate the expected value of the random variable
expected_value = np.mean(random_variable)

# Print the expected value
print("Expected Value:", expected_value)

# Create a histogram of the random variable
plt.hist(random_variable, bins=30, density=True, alpha=0.6, color='g')
plt.title('Continuous Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Calculate the theoretical expected value
theoretical_expected_value = (lower_bound + upper_bound) / 2

# Print the theoretical expected value
print("Theoretical Expected Value:", theoretical_expected_value)



# Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate two datasets with different standard deviations
dataset1 = np.random.normal(loc=0, scale=1, size=1000)
dataset2 = np.random.normal(loc=0, scale=2, size=1000)

# Calculate the standard deviations of the datasets
std_dev1 = np.std(dataset1, ddof=1)
std_dev2 = np.std(dataset2, ddof=1)

# Print the standard deviations
print("Standard Deviation of Dataset 1:", std_dev1)
print("Standard Deviation of Dataset 2:", std_dev2)

# Create histograms of the datasets
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(dataset1, bins=30, density=True, alpha=0.6, color='g')
plt.title('Dataset 1 (Standard Deviation: {:.2f})'.format(std_dev1))
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(dataset2, bins=30, density=True, alpha=0.6, color='b')
plt.title('Dataset 2 (Standard Deviation: {:.2f})'.format(std_dev2))
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

plt.tight_layout()
plt.show()

# Compare the standard deviations
if std_dev1 < std_dev2:
    print("Dataset 1 has a smaller standard deviation.")
elif std_dev1 > std_dev2:
    print("Dataset 1 has a larger standard deviation.")
else:
    print("Both datasets have the same standard deviation.")



# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate a dataset from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate the range of the dataset
data_range = np.ptp(data)

# Calculate the interquartile range (IQR) of the dataset
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25

# Print the range and IQR
print("Range:", data_range)
print("Interquartile Range (IQR):", iqr)

# Create a histogram of the dataset
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution Dataset')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Create a box plot of the dataset
plt.boxplot(data)
plt.title('Box Plot of Normal Distribution Dataset')
plt.show()


# Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate a sample dataset
data = np.random.normal(loc=10, scale=5, size=1000)

# Calculate the mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Implement Z-score normalization
z_scores = (data - mean) / std_dev

# Print the mean and standard deviation of the original and normalized datasets
print("Original Dataset:")
print("Mean:", mean)
print("Standard Deviation:", std_dev)

print("\nNormalized Dataset:")
print("Mean:", np.mean(z_scores))
print("Standard Deviation:", np.std(z_scores))

# Visualize the transformation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='b')
plt.title('Normalized Dataset (Z-Scores)')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.grid(True)

plt.tight_layout()
plt.show()



# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    # Calculate skewness
    skewness = stats.skew(data)
    
    # Calculate kurtosis
    kurtosis = stats.kurtosis(data)
    
    return skewness, kurtosis

# Set the seed for reproducibility
np.random.seed(0)

# Generate a dataset from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print the results
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

# Create a histogram of the dataset
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution Dataset')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

```
